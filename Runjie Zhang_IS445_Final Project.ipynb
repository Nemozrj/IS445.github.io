## 4. Context Visualization1 - Visualize Universities by Country

```python
# find number of countries in QS Ranking
'Number of countries with ranked universities: ',uni['country'].nunique()
```

There 97 countries that have universities in the QS World University Top1000 Rankings 2017 - 2022.
Since there are a large number of countries to visualize, I narrowed down the top1000 ranking to top100 ranking.

```python
# norrow down the dataset
uni100_2017 = uni[uni.year == 2017].iloc[:100, :]
uni100_2018 = uni[uni.year == 2018].iloc[:100, :]
uni100_2019 = uni[uni.year == 2019].iloc[:100, :]
uni100_2020 = uni[uni.year == 2020].iloc[:100, :]
uni100_2021 = uni[uni.year == 2021].iloc[:100, :]
uni100_2022 = uni[uni.year == 2022].iloc[:100, :]
uni100 = uni100_2017.append([uni100_2018, uni100_2019, uni100_2020, uni100_2021, uni100_2022])
```

```python
uni100
```

```python
count = uni100['country'].value_counts()
```

```python
count.index
```

```python
# Grammer for the visualization
# define the plot
fig, ax =plt.subplots(figsize=(100,200), dpi=150)
# define the countplot
sns.countplot(data=uni100, y='country', hue ='year', order=count.index);
# define the xlabel
plt.xlabel('Number of universities', fontsize=200)
# define the xticks
plt.xticks(fontsize=100)
# define the ylabel
plt.ylabel('Country', fontsize=200)
# define the yticks
plt.yticks(fontsize=100)
# define the legend
plt.rc('legend', fontsize=200)
# define the title
plt.title("Distribution of universities across countries", fontsize=300);
```

## 5. Central Visualization
## - Visualize University by  year interactively

```python
# find the unique year in the dataset
Year = uni100.year.unique()
```

```python
Year
```

```python
uni['year']
```

```python
# define the interact part

# get a dropdown menu lists all the years
@interact(year = Year)
def f(year):
    uni_year = uni100[uni100['year'] == (year)].groupby('country').count().reset_index()
#     print (df2)
    plt.figure(figsize = (30,50))
    ax1 = sns.barplot(y="country", x = 'university', data=uni_year)
    sns.set(font_scale = 5)
    ax1.set_xlabel("Number of University in QS Rank 100")
    ax1.set_ylabel("Country")
#     ax1.xaxis.label.set_size(50)
#     ax1.yaxis.label.set_size(50)
    plt.show()
```

### Information about the dashboard
This visualization deals with the QS World University ranking top1000 2017-2022 by year, since the number of countries in the list was large, I norrow down the ranking to top100 2017-2022.
In this visualization, it has 100 data samples every year. Grouped by country, I want to explore how the number of universities in the top100 list is different from countries. It would be clear to see what countries have a strong university education, and which countries have the top universities.
Also, there is a dropdown menu, where users can use to switch the years from 2017 to 2022. With that, compared to the countplot above in part4, users could see the information in different years more clearly.
From the dashboard, I observed that the United States has the largest number of universities in the top100, second to the United States is the United Kingdom. These two accounted for almost half of the list. Following are Australia, Japan, Canada, China(Mainland), Hong Kong SAR, South Korea, and Switzerland.

## 6. Context visualization  1
## - Visualize University by country intereactively

```python
Country = uni.country.unique()
```

```python
#define the interact part

# get a dropdown menu lists all the countries
@interact(country = Country)
def f(country):
    uni_country = uni[uni['country'] == (country)].groupby('year').count().reset_index()
#     print (df2)
    plt.figure(figsize = (30,50))
    ax1 = sns.barplot(x="year", y = 'university', data=uni_country)
    sns.set(font_scale = 5)
    ax1.set_xlabel("Number of University in QS Rank 1000")
    ax1.set_ylabel("year")
#     ax1.xaxis.label.set_size(50)
#     ax1.yaxis.label.set_size(50)
    plt.show()
```

### Information about the dashboard  
This visualization deal with the QS World University ranking top1000 2017-2022 by year. In the visualization, it has 1000 data samples every year. Grouped by year, I want to explore how the number of universities in the top1000 list different from years for one specific country. It would be clear to see changes in the ranking of universities in a country. Also, there is a dropdown menu, where users can use to switch the countries. 

### Analysis
According to Context visualization1, we find out the countries that have the most universities in the QS top100. Following this, the context visualization 2 -Visualization University by country interactively focus on to explore the trend of the number of universities varies from year to year in the QS top1000 for those countries. Take the United States for example, from 2017 to 2022, the number of universities in the top100 has countinued reduction, while the number of universities in top1000 maintains an upward trend. ## 8. Conclusion

From the above analysis, it can be concluded that the U.S. still leads college rankings, however, Asia countries like China rising, the gap is narrowing. As Ben Sowter, research director at QS said, while it is going to be “some time” before the US’s dominance as a higher education research nation could reasonably be said to be in dispute, it is being diminished by increasingly competitive nations around the world (Atack, 2018).
For China, it can be concluded that the number of universities in the top1000 has a big growth. Also, most of the Chinese universities rose their rankings. It may be attributed to its Double First-Class University Plan, under which, China’s universities have invested more in competitive US and UK-based researchers to aid in their education ("Chinese universities for international students overtake US in QS World University Rankings - Scholarships for Africans", 2021).
Also, the Covid-19 may exacerbate the trend, as China has long been the biggest source of international students at U.S. institutions. Since the full effect of Covid-19 on higher education may take years to become apparent, it may herald the start a dramatic re-balancing of the global university field (Morrison, 2020).
